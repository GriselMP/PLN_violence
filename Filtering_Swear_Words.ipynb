```python
#filtering tweets that contain words from the swear words for insult women

swear_words=["gorda","india","sirvienta","puta","zorra","perra","retrasada","in√∫til","virgen","machorra",
"est√∫pida","naca","idiota","teibolera","vedette","gata","mierda","malparida","obesa","vendida",
"prostituta","vaca","cerda","golfa","chaquetera","feminazi","fea","tetas","asesina","arrastrada",
"robamaridos","bastarda","maldita","escoria","tonta","miserable","torpe","pendeja","coger","puta madre",
"culo","mamahuevos","lame-botas","boba","pinche","panocha","feminismo","pobre","negra","se√±ora",
"bruja","arp√≠a","fodonga","vieja","asquerosa","chismosa","mentirosa","dram√°tica","perdedora","ramera","hija de puta",
"mojigata","barata","hist√©rica","neur√≥tica","solterona","ama de casa","mantenida","imb√©cil","lesbiana","baja novios",
"mujeres","mujer","zopilota","actriz"]

def filter_swear_words(lista1,lista2):
  out=[]
  for elem in lista1:
    if elem in lista2:
      out=lista1
  return out

#read all data extracted from Twitter with the script
f = open("/content/drive/MyDrive/basesTwitter/dataset_spaces.txt", "r")
tuits=f.readlines()
f.close()

#extract all tweets that contain swear words and save into a txt file
fi = open("/content/drive/MyDrive/basesTwitter/candidatosV.txt", "w")
for tuit in tuits:
    frase=tuit.split()
    vi=filter_swear_words(frase,swear_words)
    filaV=" ".join(vi)

    fi.write(filaV+"\n")
fi.close()
```
